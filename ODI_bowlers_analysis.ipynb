#### rename the column names

```python
print(df.columns)
```

```python
#rename the column names
df=df.rename(columns={'Mat':'Match',
                     'Inns':'Innings',
                     'Wkts':'Wickets',
                     'Ave':'Average',
                     'Econ':'Economy',
                     'SR':'Strike_Rate',
                     4:'Four',
                     5:'Five'
                     })
display(df.head())
```

#### players played for ICC

```python
#spilt the player columns 
df_player =df['Player'].str.split("(", expand=True)
display(df_player.head())
```

```python
#name the columns 
df[["Bowler_name", "Country"]]=df_player
display(df.head())
```

```python
#drop unwanted columns 
df=df.drop('Player', axis=1)
display(df.head())
```

```python
#replace value from columns 
df["Country"]=df["Country"].str.replace(')','')
display(df.head())
```

```python
#print columns name
print(df.columns)
```

```python
new_col_sequence=['Bowler_name','Country','Span', 'Match', 'Innings', 'Balls', 'Runs', 'Wickets', 'Average',
       'Economy', 'Strike_Rate', 'Four', 'Five']
```

```python
df = df[new_col_sequence]

display(df.head())
```

```python
#find number of country bowler each
df['Country'].value_counts()
```

```python
def icc_check(x):
    if "ICC" in x:
        return "Yes"
    else:
        return "No"
```

```python
def asia_check(x):
    if "Asia" in x:
        return "Yes"
    else:
        return "No"
```

```python
def africa_check(x):
    if "Afr" in x:
        return "Yes"
    else:
        return "No"
```

```python
df['played_for_ICC'] = df['Country'].apply(icc_check)
display(df.head())
```

```python
#player for ICC 
df['played_for_ICC'].value_counts()
```

```python
df['played_for_Asia'] = df['Country'].apply(icc_check)
df['played_for_Afr'] = df['Country'].apply(icc_check)
display(df.head())
```

```python
#remove/replace unwanted thing from country 
df['Country']=df['Country'].str.replace('Asia/','')
df['Country']=df['Country'].str.replace('ICC/','')
df['Country']=df['Country'].str.replace('/ICC','')
df['Country']=df['Country'].str.replace('Afr/','')
display(df.head())
```

#### number of different countries present

```python
#countries 
df['Country'].value_counts()
df['Country'].value_counts()
```

#### played for the longest

```python
#split span columns 
df[['start_year', 'end_year']] = df['Span'].str.split("-", expand=True)
display(df.head())
```

```python
#drop Span columns
df = df.drop("Span", axis=1)
display(df.head())
```

```python
df.info()
```

```python
#data type change 
df['start_year'] = df['start_year'].astype('int') 
df['end_year'] = df['end_year'].astype('int')
#df.info()
print(df.info())
```

```python
df['years_played'] = df['end_year'] - df['start_year']
display(df.head())
```

```python
# played for the longest period of time
#df[['years_played', 'Bowler_name']].head()
#display(df.head())
df.sort_values(by='years_played', ascending = False).head()
```

#### played for the shortest period of time

```python
#played for the shortest period of time
df.sort_values(by='years_played', ascending = True).head()
```
